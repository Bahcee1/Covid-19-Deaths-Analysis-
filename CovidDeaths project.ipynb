# ANALYSIS

```python
#1.impact of population density on COVID-19 spread
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
correlation_matrix = df[['population_density', 'total_cases_per_million', 'new_cases_per_million',
                            'total_deaths_per_million']].corr()
print(correlation_matrix)
```

```python
# Scatter plot for Population Density vs Total Cases per Million
sns.scatterplot(x='population_density', y='total_cases_per_million', data=df)
plt.title('Population Density vs Total Cases per Million')
plt.xlabel('Population Density (people per square km)')
plt.ylabel('Total Cases per Million')
plt.show()

# Scatter plot for Population Density vs New Cases per Million
sns.scatterplot(x='population_density', y='new_cases_per_million', data=df)
plt.title('Population Density vs New Cases per Million')
plt.xlabel('Population Density (people per square km)')
plt.ylabel('New Cases per Million')
plt.show()

# Scatter plot for Population Density vs Total Deaths per Million
sns.scatterplot(x='population_density', y='total_deaths_per_million', data=df)
plt.title('Population Density vs Total Deaths per Million')
plt.xlabel('Population Density (people per square km)')
plt.ylabel('Total Deaths per Million')
plt.show()
```

```python
#2. Relationship between the Stringency Index and COVID-19 cases or deaths
```

```python
stringency = covid.dropna(subset=["stringency_index", "new_cases", "new_deaths"])
```

```python
stringency[["stringency_index", "new_cases", "new_deaths"]].corr()
```

```python
#scatter plot for strigency index vs new cases

sns.scatterplot(x="stringency_index",y="new_cases", data=stringency)
plt.title('Stringency Index vs New Cases')
plt.xlabel('Stringency Index')
plt.ylabel('New Cases')
plt.show()

#scatter plot for strigency index vs new deaths

sns.scatterplot(x="stringency_index",y="new_deaths", data=stringency)
plt.title('Stringency Index vs New Deaths')
plt.xlabel('Stringency Index')
plt.ylabel('New Deaths')
plt.show()

"""These plots will provide a visual representation of the relationship between the Stringency Index and COVID-19 outcomes. 
If the data points trend downward as the Stringency Index increases, 
it suggests that stricter measures were associated with fewer cases or deaths."""
```

```python
cov=covid.dropna(subset=["continent", "total_deaths"])
```

```python
#3. Total Cases and Deaths per Continent

cov.groupby(["continent"]).sum(numeric_only=True)["total_deaths"].reset_index()
```

```python
sns.barplot(x="continent",y="total_deaths",data=cov)
plt.title("Total Deaths by Continent")
plt.xlabel("CONTINENT")
plt.ylabel("TOTAL DEATHS")
plt.show()
```

```python
#4. correlation between age demographics and COVID-19 severity
```

```python
 age[['median_age', 'aged_65_older', 'total_deaths_per_million']].corr()

"""The strong correlation between median_age and aged_65_older confirms that as the median age increases,
so does the proportion of the population aged 65 and older.
Both median_age and aged_65_older have moderate positive correlations with total_deaths_per_million. 
This suggests that age demographics do play a role in COVID-19 severity,
with older populations experiencing higher death rates, but other factors are likely at play as well."""
```

```python
# Scatter plot for Median Age vs Total Deaths per Million
sns.scatterplot(x='median_age', y='total_deaths_per_million', data=age)
plt.title('Median Age vs Total Deaths per Million')
plt.xlabel('Median Age')
plt.ylabel('Total Deaths per Million')
plt.show()

# Scatter plot for Aged 65 and Older vs Total Deaths per Million
sns.scatterplot(x='aged_65_older', y='total_deaths_per_million', data=age)
plt.title('Aged 65 and Older vs Total Deaths per Million')
plt.xlabel('Proportion Aged 65 and Older')
plt.ylabel('Total Deaths per Million')
plt.show()
```
